# Opening and visualizing images

Most images with standard extensions (tif, png etc.) can be read using the [```skimage.io.imread```](https://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imread) function.

See also
* [Opening other file formats using Python](https://haesleinhuepf.github.io/BioImageAnalysisNotebooks/12a_image_file_formats/readme.html)

To use ```imread``` you have three possibilities:
- use the absolute path to an image file e.g. ```imread('/Users/username/Desktop/blobs.tif')```
- use a relative path to where you currently are, e.g. ```imread('data/blobs.tif')``` or ```imread('../../data/blobs.tif')```
- use a url that points to an image file, e.g. from a GitHub repository ```imread('https://github.com/haesleinhuepf/BioImageAnalysisNotebooks/raw/main/data/blobs.tif')```

Here we use a relative path. Respective to the current notebook, the data in a sub-folder called ```data```:

````python
from skimage.io import imread
from matplotlib.pyplot import imshow
import stackview

image = imread("data/mitosis_mod.tif")
image
````

## Displaying images

There are many ways to display simple 2D images. In many notebooks and examples online, you will find examples using Matplotlib's ```imshow``` function:

````python
imshow(image)
````

# Alternative: Stackview
There are some alternatives which fit needs in scientific / microscopy image analysis better, e.g. [stackview](https://github.com/haesleinhuepf/stackview).

````python
import stackview
stackview.imshow(image)
````

Stackview can also tell us more about the image.

````python
stackview.insight(image)
````

## Lookup tables (a.k.a. color maps)
We can also change the [color map](https://matplotlib.org/stable/tutorials/colors/colormaps.html), a.k.a. "lookup table" for the visualization.

````python
stackview.imshow(image, colormap="pure_red")
````

````python
stackview.imshow(image, colormap="seismic")
````
